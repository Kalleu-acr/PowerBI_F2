```python
# FONTE: https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/saeb
# SAEB

# BIBLIOTECAS
import pandas as pd
import numpy as np
from google.colab import files

# UPLOAD CSV
print("Realize o upload do arquivo do SAEB (CSV)")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# LEITURA DO CSV


# Colunas categóricas (ajustadas ao padrão SAEB)
cat_cols = [
    'ID_REGIAO',
    'ID_UF',
    'IN_PUBLICA',
    'ID_LOCALIZACAO',
    'ID_AREA',
    'NIVEL_SOCIO_ECONOMICO'
]

df = pd.read_csv(
    file_name,
    sep=';',
    encoding='latin1',
    low_memory=False
)

print("Arquivo CSV lido com sucesso")
print("Shape inicial:", df.shape)


# Padronização de nomes de colunas
df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(" ", "_")
)

print("\n Colunas disponíveis no dataset:")
print(list(df.columns))

# Checar colunas críticas esperadas
colunas_esperadas = ['ID_ESCOLA', 'ID_UF']

colunas_faltantes = [col for col in colunas_esperadas if col not in df.columns]

if colunas_faltantes:
    print(f"\n colunas ausentes: {colunas_faltantes}")
    print("revise o dataset.")
else:
    print("\nColunas essenciais verificadas")



for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

if "NIVEL_SOCIO_ECONOMICO" in df.columns:
    df["NIVEL_SOCIO_ECONOMICO"] = (
        df["NIVEL_SOCIO_ECONOMICO"]
        .astype(str)
        .str.replace("N韛el", "Nível", regex=False)
        .str.strip()
    )


# Vazio → NaN
df.replace("", np.nan, inplace=True)

# Codigo comum invalido
df.replace({88888: np.nan}, inplace=True)

# Substituição de 9 por NaN em colunas numéricas
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].replace(9, np.nan)

print("Valores inválidos tratados")

for col in df.columns:
    if col not in cat_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

print("Conversão de tipos concluída")

taxa_cols = [col for col in df.columns if "TAXA" in col]

if taxa_cols:
    df[taxa_cols] = df[taxa_cols].apply(pd.to_numeric, errors="coerce")

    for col in taxa_cols:
        df.loc[df[col] > 100, col] = np.nan

    print(f"{len(taxa_cols)} colunas de taxa tratadas")
else:
    print("Nenhuma coluna de taxa encontrada")


print("\n Info geral do dataset:")
print(df.info())

print("\n Top 15 colunas com mais valores nulos:")
print(df.isnull().sum().sort_values(ascending=False).head(15))

if "ID_ESCOLA" in df.columns:
    duplicados = df.duplicated(subset=["ID_ESCOLA"]).sum()
    print(f"\n Registros duplicados por escola: {duplicados}")

output_file = "saeb_tratado.csv"
df.to_csv(output_file, index=False)

print(f"\nArquivo salvo como {output_file}")

files.download(output_file)

print("\n Processo concluido")
```
